```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

```python
rent = pd.read_csv("FINAL_RENT.csv")
rent
```

```python
# center to the mean of all points
m = folium.Map(location=rent[["Lat", "Lon"]].mean().to_list(), zoom_start=11)

# draw the markers and assign popup and hover texts
# add the markers the the cluster layers so that they are automatically clustered
for i,r in rent.iterrows():
    location = (r["Lat"], r["Lon"])
    folium.Marker(location=location,
                  popup = r['Name'] + '\n$' + str(r['Price']),
                  tooltip=r['Name'])\
    .add_to(m)

# display the map
m
```

```python
m.save("folium_map_rent.html")
```

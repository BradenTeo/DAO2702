```python
area_sqft = rent['property_area_or_psf'].str.split(r" ", n=-1, expand=True)

area_sqft[5] = area_sqft[0].str.replace(",", "")
area_sqft['area (sqft)'] = area_sqft[5].str[:].astype(int)

area_sqft['psf price ($/psf)'] = area_sqft[3].str[1:].astype(float)
area_sqft.drop([0, 1, 2, 3, 4, 5], axis=1, inplace=True)

print(area_sqft.dtypes)
area_sqft
```

```python
parsed_rent = pd.concat([rent, area_sqft], axis=1)
parsed_rent.drop(['property_area_or_psf'], axis=1, inplace=True)
parsed_rent
```

```python
parsed_rent['total price'] = parsed_rent['property_total_price'].str.replace(",", "")
parsed_rent['total price ($)'] = parsed_rent['total price'].str[1:].astype(int)
parsed_rent.drop(['property_total_price', 'total price'], axis=1, inplace=True)

print(parsed_rent.dtypes)
parsed_rent
```

```python
condition = (parsed_rent['area (sqft)'] <= 2000) & (parsed_rent['area (sqft)'] >= 1500)
rent_satisfy_area = parsed_rent.loc[condition]
final_rent = rent_satisfy_area.sort_values('psf price ($/psf)')
final_rent.reset_index(drop=True, inplace=True)
final_rent.to_csv('updated_rent.csv')
final_rent
```
